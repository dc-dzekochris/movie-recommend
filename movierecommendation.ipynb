```python
# Users Datasets
import pandas as pd
users = pd.read_csv('https://msi.martial.co.ke/data/file.csv')
users
```

```python
# Movie IDS Dataset
movies = pd.read_csv('https://msi.martial.co.ke/data/Movie_Id_Titles.csv')
movies
```

```python
#join two data sets
# Merge the users and movies Dataset
df = pd.merge(users,movies, on='item_id')
df
```

```python
# Exploratory Data Analysis
# 1. Information
df.info()
```

```python
# 2. Maximium and Minimum Rating and Avearage Rating
df['rating'].max()
df['rating'].min()
df['rating'].mean()
df['title'].count()
```

```python
# 3. Statistical Analysis
# Total Users: 943
# Total Movies: 100003
df.describe()
```

```python
# 4.Checking the Empties
df.isnull().sum()
```

```python
#grouping all movie types according to the avaerage rating they have
mean_rating=df.groupby('title')['rating'].mean()

#highest rated movies of IMDB
mean_rating.sort_values(ascending=False).head(10)
```

```python
#least 5 rated movies
mean_rating.sort_values(ascending=True).head(5)
```

```python
#popularity of a movie
rating_count = df.groupby('title')['rating'].count()
rating_count.sort_values(ascending=False).head(10)
```

```python
#least popular movie
rating_count = df.groupby('title')['rating'].count()
rating_count.sort_values(ascending=False).head(10)
```

```python
# Create a DataFrame that will Contain, the title, mean_rating, number_of_views
df_new = pd.DataFrame(mean_rating)
df_new['number_of_views'] = pd.DataFrame(rating_count)
df_new.sort_values(by='number_of_views', ascending=False).head(10)
```

```python
# PIVOT TABLE
# Data can be Viewed in 3D
# X-Columns
# Y-indexing
# Z-values
```

```python
pd.set_option('display.max_columns', 100)
pivot = df.pivot_table(index='user_id', columns='title', values='rating')
pivot
```

```python
selected_movie = pivot['Godfather, The (1972)']
selected_movie
```

```python
#recommend a similar movie
#correlation: needs to have a good positive correlation with the selected movie
#ie. if the rating and users increase in selected movies, it should also embrace in similar movie

import warnings
warnings.filterwarnings('ignore')

similar = pivot.corrwith(selected_movie)
similar.sort_values(ascending=False).head(10)
```

```python
similar = pivot.corrwith(selected_movie)
similar.sort_values(ascending=False).head(10)

similar_df = pd.DataFrame(similar, columns=['correlation'])
similar_df.sort_values('correlation', ascending=False).head(10)
```

```python

# Join the similar_df with the number of Views
similar_df = similar_df.join(df_new['number_of_views'])
similar_df.head(10)
```

```python
# Recommendation:
# Positive Correlation
# Higher Number of Views
condition = similar_df['number_of_views'] > 200
similar_df = similar_df[condition]
similar_df[condition].sort_values('correlation', ascending=False).head(10)

#condition = similar_df['number_of_views'] > 200
#similar_df = similar_df[condition]
#recommended_movies = similar_df.sort_values(by='Correlation', ascending=False)
#recommended_movies.head(10)
```
